Library Needs

```python
from PIL import Image, ImageOps
import matplotlib
import matplotlib.pyplot as plt
from pylab import *
```

Adding Border Around Image

```python
# Function used to accepts arguments for the paths of input image, output image and border 
def add_border(input_image, output_image, border):
    img = Image.open(input_image)
    
    '''  Draws a black border around an image. The output image will
    have a squared image keeping the ratio of the input image.'''  
    
    if isinstance(border, int) or isinstance(border, tuple):
        bimg = ImageOps.expand(img, border=border)
    else:
        raise RuntimeError('Border is not an integer or tuple!')
 
    bimg.save(output_image)
 
if __name__ == '__main__':
    in_img = 'balloon.jpg'
 
    add_border(in_img, output_image='balloon_border.jpg',
               border=200)
```

Change Border And Color Around Image  

```python
#Function accepts an additional argument new color 
def add_border(input_image, output_image, border, color = 0):
    img = Image.open(input_image)
    
    '''  Draws a thickened black border around an image.'''  
    
    if isinstance(border, int) or isinstance(border, tuple):
        bimg = ImageOps.expand(img, border=border, fill=color)
    else:
        raise RuntimeError('Border is not an integer or tuple!')
 
    bimg.save(output_image)
 
if __name__ == '__main__':
    in_img = 'balloon.jpg'
 
    add_border(in_img,
               output_image='balloon_border_orange_top_bottom.jpg',
               border=(5,100),color='orange')
```

Plot Image With Graph Coordinates

```python
# Read in image as an array
im = array(Image.open('balloon.jpg'))

# Plot the image
imshow(im)

# Add title and show the plot
title('Plotting: "balloon.jpg"')
show()
```

Crop Image 

```python
# Read in image 
image = Image.open('balloon.jpg')

# Select region to crop
box = (100, 100, 600, 300)
croppedIm = image.crop(box)

# Save cropped image
croppedIm.save('cropped.jpg')

# Plot the image
imshow(croppedIm)

axis('off')

# Add title and show the plot
title('Plotting: "cropped.jpg"')
show()
```

Create Histogram from Grayscale Image  

```python
# Read image in as array and change to gray scale
im = array(Image.open('balloon.jpg').convert('L'))

# Show contours with origin upper left corner
contour(im, origin='image')
axis('equal')
axis('off')

# Add title and plot
title('Plotting: "Contour Visual"')

# Create new image and plot a histogram of 
figure()
hist(im.flatten(),133)

# Add title and show graph
title('Plotting: "Grayscale Histogram"')
plt.ylabel("# of Pixels")
plt.xlabel("Bins")

# Plot histogram 
show()
```
